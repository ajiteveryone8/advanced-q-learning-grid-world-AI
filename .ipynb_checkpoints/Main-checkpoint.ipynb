```python
import numpy as np
import random

# Parameters
n_states = 5
actions = [0, 1]  # 0 = left, 1 = right
q_table = np.zeros((n_states, len(actions)))

learning_rate = 0.1
discount_factor = 0.9
epsilon = 0.3  # exploration rate
episodes = 100

for episode in range(episodes):
    state = 0
    while state != 4:
        # ε-greedy strategy
        if random.uniform(0, 1) < epsilon:
            action = random.choice(actions)
        else:
            action = np.argmax(q_table[state])

        # Take action
        next_state = state - 1 if action == 0 else state + 1
        next_state = max(0, min(next_state, n_states - 1))

        # Reward
        reward = 1 if next_state == 4 else 0

        # Update Q-table
        q_table[state][action] = q_table[state][action] + learning_rate * (
            reward + discount_factor * np.max(q_table[next_state]) - q_table[state][action]
        )

        state = next_state

# Show final Q-table
print("Q-table after training:")
print(q_table)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Grid size
rows, cols = 5, 5
states = [(i, j) for i in range(rows) for j in range(cols)]
actions = ['up', 'down', 'left', 'right']
q_table = {state: np.zeros(len(actions)) for state in states}

# Hyperparameters
lr = 0.1
gamma = 0.9
epsilon = 0.3
episodes = 200

# Helper: Get next state
def move(state, action):
    i, j = state
    if action == 'up':
        i = max(i - 1, 0)
    elif action == 'down':
        i = min(i + 1, rows - 1)
    elif action == 'left':
        j = max(j - 1, 0)
    elif action == 'right':
        j = min(j + 1, cols - 1)
    return (i, j)

# Training
for ep in range(episodes):
    state = (0, 0)
    while state != (4, 4):
        if random.random() < epsilon:
            action_idx = random.randint(0, 3)
        else:
            action_idx = np.argmax(q_table[state])
        action = actions[action_idx]
        next_state = move(state, action)
        reward = 10 if next_state == (4, 4) else -1
        q_table[state][action_idx] += lr * (reward + gamma * np.max(q_table[next_state]) - q_table[state][action_idx])
        state = next_state

# 💡 Display policy
policy_grid = np.empty((rows, cols), dtype=str)
arrow_map = ['↑', '↓', '←', '→']

for state in states:
    if state == (4, 4):
        policy_grid[state] = '🏁'
    else:
        best_action = np.argmax(q_table[state])
        policy_grid[state] = arrow_map[best_action]

print("\nPolicy Grid:")
print(policy_grid)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random

# Grid size
rows, cols = 5, 5
states = [(i, j) for i in range(rows) for j in range(cols)]
actions = ['up', 'down', 'left', 'right']
q_table = {state: np.zeros(len(actions)) for state in states}

# Hyperparameters
lr = 0.1
gamma = 0.9
epsilon = 0.3
episodes = 200

# Action helper
def move(state, action):
    i, j = state
    if action == 'up':
        i = max(i - 1, 0)
    elif action == 'down':
        i = min(i + 1, rows - 1)
    elif action == 'left':
        j = max(j - 1, 0)
    elif action == 'right':
        j = min(j + 1, cols - 1)
    return (i, j)

def print_q_table(q_table):
    for row in range(5):
        for col in range(5):
            state = (row, col)
            q_values = q_table[state]
            best_action = np.argmax(q_values)
            action_symbols = [' ']*4
            action_symbols[best_action] = actions[best_action]
            print(f"[{''.join(action_symbols)}]", end=" ")
        print()
    print("\n")

# Training
for ep in range(episodes):
    print("\n")
    print_q_table(q_table)
    state = (0, 0)
    steps = 1
    while state != (4, 4):
        if random.random() < epsilon:
            action_idx = random.randint(0, 3)
        else:
            action_idx = np.argmax(q_table[state])
        action = actions[action_idx]
        next_state = move(state, action)
        reward = 10 if next_state == (4, 4) else -1
        q_table[state][action_idx] += lr * (reward + gamma * np.max(q_table[next_state]) - q_table[state][action_idx])
        
        # Show the step
        print(f"Episode {ep} -  Step {steps} -  {state} --{action}--> {next_state} | reward={reward}")
        steps +=1
        state = next_state

# Get optimal path for animation
def get_optimal_path(start):
    path = [start]
    current = start
    while current != (4, 4):
        best_action = np.argmax(q_table[current])
        current = move(current, actions[best_action])
        path.append(current)
        if len(path) > 50:  # Safety check
            break
    return path

path = get_optimal_path((0, 0))

# Visualization
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-0.5, cols - 0.5)
ax.set_ylim(-0.5, rows - 0.5)
ax.set_xticks(np.arange(-0.5, cols, 1))
ax.set_yticks(np.arange(-0.5, rows, 1))
ax.grid(True)
ax.invert_yaxis()
agent_dot, = ax.plot([], [], 'ro', markersize=20)
goal = plt.Rectangle((4 - 0.5, 4 - 0.5), 1, 1, color='green', alpha=0.3)
ax.add_patch(goal)

def init():
    agent_dot.set_data([], [])
    return agent_dot,

def update(frame):
    pos = path[frame]
    agent_dot.set_data(pos[1], pos[0])
    return agent_dot,

ani = animation.FuncAnimation(
    fig, update, frames=len(path), init_func=init,
    interval=500, blit=True, repeat=False
)



plt.title("Agent Moving to Goal (Q-Learning)")
plt.show()
```
